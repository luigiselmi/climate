# FaIR Model
FaIR stands for Finite Amplitude Impulse Response simple climate model.

```julia
from fair import FAIR
f = FAIR()
```

```julia
from fair.io import read_properties
f.define_time(2000, 2050, 1)
print(f.timebounds)
print(f.timepoints)
```

```julia
f.define_scenarios(['abrupt', 'ramp'])
f.scenarios
```

```julia
f.define_configs(['high', 'central', 'low'])
f.configs
```

```julia
data_url = 'https://raw.githubusercontent.com/OMS-NetZero/FAIR/refs/heads/master/examples/data/importing-data/species_configs_properties.csv' 
species, properties = read_properties(data_url)
f.define_species(species, properties)
f.ghg_method = 'Myhre1998'
```

```julia
f.species
```

```julia
f.properties
```

```julia
f.allocate()
```

```julia
f.emissions
```

```julia
f.temperature
```

```julia
f.emissions.loc[(dict(specie='CO2 FFI', scenario='abrupt'))] = 38
```

```julia
emission_url = 'https://raw.githubusercontent.com/OMS-NetZero/FAIR/refs/heads/master/examples/data/basic_run_example/emissions.csv'
concentration_url = 'https://raw.githubusercontent.com/OMS-NetZero/FAIR/refs/heads/master/examples/data/basic_run_example/concentration.csv'
forcing_url = 'https://raw.githubusercontent.com/OMS-NetZero/FAIR/refs/heads/master/examples/data/basic_run_example/forcing.csv'
f.fill_from_csv(
    emissions_file = emission_url,
    concentration_file = concentration_url,
    forcing_file = forcing_url
)
```

```julia
import pandas as pd
pd.read_csv(emission_url)
```

```julia
pd.read_csv(concentration_url)
```

```julia
pd.read_csv(forcing_url)
```

```julia
from fair.interface import initialise
initialise(f.concentration, 278.3, specie='CO2')
initialise(f.forcing, 0)
initialise(f.temperature, 0)
initialise(f.cumulative_emissions, 0)
initialise(f.airborne_emissions, 0)
initialise(f.ocean_heat_content_change, 0)
```

```julia
f.fill_species_configs(data_url)
```

```julia
f.species_configs
```

```julia
ensemble_url = 'https://raw.githubusercontent.com/OMS-NetZero/FAIR/refs/heads/master/examples/data/basic_run_example/configs_ensemble.csv'
f.override_defaults(ensemble_url)
```

```julia
pd.read_csv(ensemble_url)
```

```julia
f.run()
```

```julia
import matplotlib.pyplot as pl
pl.plot(f.timebounds, f.temperature.loc[dict(scenario='ramp', layer=0)], label=f.configs)
pl.title('Ramp scenario: temperature')
pl.xlabel('year')
pl.ylabel('Temperature anomaly (K)')
pl.legend()
```

```julia
pl.plot(f.timebounds, f.concentration.loc[dict(scenario='ramp', specie='CO2')], label=f.configs)
pl.title('Ramp scenario: CO2')
pl.xlabel('year')
pl.ylabel('CO2 (ppm)')
pl.legend()
```

```julia
pl.plot(f.timebounds, f.forcing.loc[dict(scenario='ramp', specie='Aerosol-cloud interactions')], label=f.configs)
pl.title('Ramp scenario: forcing')
pl.xlabel('year')
pl.ylabel('ERF from aerosol-cloud interactions (W m$^{-2}$)')
pl.legend()
```

```julia
pl.plot(f.timebounds, f.temperature.loc[dict(scenario='abrupt', layer=0)], label=f.configs)
pl.title('Abrupt scenario: temperature')
pl.xlabel('year')
pl.ylabel('Temperature anomaly (K)')
pl.legend()
```

```julia
pl.plot(f.timebounds, f.forcing_sum.loc[dict(scenario='abrupt')], label=f.configs)
pl.title('Abrupt scenario: forcing')
pl.xlabel('year')
pl.ylabel('Total ERF (W m$^{-2}$)')
pl.legend()
```

```julia
pl.plot(f.timebounds, f.concentration.loc[dict(scenario='abrupt', specie='CO2')], label=f.configs)
pl.title('Abrupt scenario: CO2')
pl.xlabel('year')
pl.ylabel('CO2 (ppm)')
pl.legend()
```
